### Le Web
---

Le ***Web***, abréviation de l'anglais ***World Wide Web*** (littéralement "toile d'envergure mondiale") est un ensemble de documents liés entre eux au moyen de ***liens hypetertexte*** : ces documents peuvent être parcourus de façon non linéaire en "sautant" d'un document à l'autre de manière interactive en cliquant sur ces liens.

Cette infrastructure s'appuie sur le ***réseau Internet***, `HTTP` étant l'un des nombreux protocoles utilisés de la couche `Application` s'exécutant au- dessus des protocoles `TCP` et `IP`.

### Les Langages et leur Rôle
---

Par extension, on englobe aussi sous le terme de Web toute l'infrastrucure hébergeant cette collection mondiale de documents, à savoir :
- Les langages permettant d'écrire les documents hypertextes : `HTML` et `CSS`
- L'architecture client-serveur utilisant le protocole `HTTP` ou sa version sécurisée `HTTPS`
- Des langages de programmation côté serveur dont le but est de générer du `HTML` : `PHP`, `PYTHON`, `Ruby`, ...
- Des langages de programmation côté client pour permettre l'écriture d'applications Web interactives : `JavaScript`

Ce chapitre a pour but de présenter les langages `HTML`, `CSS` et `JS`, prérequis nécessaire pour introduire les interactions client-serveur et les interactions entre l'utilisateur et une page Web.

**<u>↪ À Retenir</u>** : Rôle respectifs des langages `HTML`, `CSS` et `JS`
> - `HTML` : Pour la ***structure*** et le ***contenu***
>   - Le contenu : Le texte, les liens hypertexte, les sons, les images, les vidéos, ...
>   - La structure : Paragraphes, listes, sections, menu de navigation, ...  
> - `CSS` : Pour la "***déco***"
> <br >Ce langage permet principalement de mettre en forme le contenu de la page `HTML` (Positionnement, agencement, couleur, police, taille de caractères, ...)
> - `JS` : Pour rendre la ***page dynamique***
> <br >Ce langage de programmation permet de rendre la page interactive

### Structure de base d'une page en `HTML5`
---

Voici la structure type d'une page `index.html` en HTML5 ...
> ↪ *Code HTML* :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Titre Onglet</title>   
>     </head>
>     <body>
>         <p>Voici un 1er paragraphe</p>
>         <p>En voilà un second</p>
>         <!-- Le reste du contenu de ma page à insérer ici -->
>     </body>
> </html>
> ```
> <br >↪ *Rendu correspondant dans un navigateur* :

<figure>
	<center>
		
		<figcaption></figcaption>
	</center>
</figure>

Les textes en vert, entre chevrons, sont des ***balises*** : Elles indiquent la nature du texte qu'elles encadrent, comme un titre, une image, un paragraphe, un commentaire, ...

On distingue deux types de balises :   	 
- Les balises *en paires* comme <span style="color:green;">&lt;title&gt; ... &lt;/title&gt;</span>, <span style="color:green;">&lt;body&gt; ... &lt;body&gt;</span>, ... : 
  <br >La balise ouvrante <span style="color:green;">&lt;balise&gt;</span> et la balise fermante <span style="color:green;">&lt;/balise&gt;</span>.
-	Les balises *orphelines* comme <span style="color:green;">&lt;meta &gt;</span>, <span style="color:green;">&lt;img &gt;</span>, ... : 
  <br >Elles permettent le plus souvent d'insérer un élément dans un endroit précis.

Par ailleurs, une balise peut posséder des attributs qui sont en quelques sorte des options donnant des informations supplémentaires. La syntaxe est la suivante :
```html
<balise attribut1="valeur1" attribut2="valeur2"> ... </balise>
ou
<balise attribut1="valeur1" attribut2="valeur2" />
```

Analysons cette structure de plus près :
- Le ***doctype*** `<!DOCTYPE html>` : Cette ligne indique qu'il s'agit bien d'une page web `HTML`.
- La balise <span style="color:green;">&lt;html&gt; ... &lt;/html&gt;</span> : C'est la balise principale du code. Elle englobe tout le contenu de la page.
- L'en-tête <span style="color:green;">&lt;head&gt; ... &lt;/head&gt;</span> : Elle donne quelques informations générales sur la page comme :
  - L'encodage des caractères : <span style="color:green;">&lt;meta&gt;</span> <span style="color:#1969a2;">charset</span>=<span style="color:red;">"utf-8"</span> <span style="color:green;">&gt;</span><br >***Attention !*** Il ne suffit pas de dire que le fichier est encodé en `UTF-8` : Il faut aussi qu'il soit bien enregistré en `UTF-8`.
  - Le titre de la page qui s'affiche dans l'onglet : <span style="color:green;">&lt;title&gt; ... &lt;/title&gt;</span>
- Le corps <span style="color:green;">&lt;body&gt; ... &lt;/body&gt;</span> : Il contient l'ensemble du contenu de la page Web.
- Les commentaires <span style="color:#408080;">&lt;!-- ... --&gt;</span> : Les textes mis en commentaires ne sont pas lus par les navigateurs. Ils permettent de laisser des indications sur le fonctionnement de la page créée, ce qui est utile pour une personne qui va lire le code source de la page.

*Exemple n°1*:
> Créez votre première page en respectant les contraintes suivantes :
> - Enregistrer la page sous `first_page.html`
> - Titre de la page : `My Very First Page`
> - Encodage : `UTF-8`
> - Texte à écrire dans un *seul paragraphe* (balises <span style="color:green;">&lt;p&gt; ... &lt;/p&gt;</span>) en respectant le *retour à la ligne*  : 
>   ```
>   Hello World ! 
>   Voici ma première page Web.
>   ```
    


### Organisation du texte

On rappelle que ***tout le contenu de la page web*** doit être ***entre les balises <span style="color:green;">&lt;body&gt; ... &lt;/body&gt;</span>***.

#### Les titres de sections
---

En `HTML`, il existe six niveaux de titre permettant de définir des titres de sections, sous-section, etc ...
> ↪ *Code HTML* :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Les titres de section</title>   
>     </head>
>     <body>
>         <h1>Titre de la page en général</h1>
>         <h2>Un titre de section</h2>
>         <h3>Un premier sous-titre de la 1ère section</h3>
>         <h3>Un second sous-titre de la 1ère section</h3>
>         <h2>Une seconde section</h2>
>         <h3>Un premier sous-titre de la 2ème section</h3>
>         <h4>Un premier sous-sous-titre de la 2ème section</h4>
>         <h5>Un premier sous-sous-sous-titre de la 2ème section</h5>
>         <h6>Un premier sous-sous-sous-sous-titre de la 2ème section</h6>
>     </body>
> </html>
> ```
> <br >↪ *Rendu correspondant* :

#### Les paragraphes
---

La plupart du temps, le texte d'une page web doit se trouver à l'intérieur d'un paragraphe. Les balises <span style="color:green;">&lt;p&gt; ... &lt;/p&gt;</span> délimitent les paragraphes. 
Ces balises étant de `type block`, un saut de ligne est effectué entre deux paragraphes.

Pour créer un simple retour à la ligne au sein d'un même paragraphe, on utilise la balise orpheline <span style="color:green;">&lt;br &gt;</span>. Celle-ci doit obligatoirement se trouver à l'intérieur d'un paragraphe.
> ↪ *Code HTML* :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Paragraphes</title>   
>     </head>
>     <body>
>         <p>Mon 1er paragraphe</p>
>         <p>
>             Un second avec un ...
>             <br >... retour à la ligne
>         </p>
>     </body>
> </html>
> ```
> <br >↪ *Rendu correspondant* :

#### Les listes
---

On peut organiser le texte en listes énumérées ou non énumérées :
- Une liste énumérée est contenue dans une balise <span style="color:green;">&lt;ol&gt; ... &lt;/ol&gt;</span> (ordered list)
- Une liste énumérée est contenue dans une balise <span style="color:green;">&lt;ul&gt; ... &lt;/ul&gt;</span> (unordered list)

Quel que soit le type de liste, chaque item est contenu dans une balise <span style="color:green;">&lt;li&gt; ... &lt;/li&gt;</span>

> ↪ ***Code HTML*** :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Liste numérotée</title>
>     </head>
>     <body>
>         <p>Mes séries</p>
>         <ol>
>             <li>Peaky Blinders</li>
>             <li>Narcos</li>
>             <li>Mr Robot</li>
>         </ol>
>     </body>
> </html>
> ```

#### Les tableaux
---

Il est également possible de créer des tableaux à deux dimensions à l'aide de la balise balises <span style="color:green;">&lt;table&gt; ... &lt;/table&gt;</span> :
- Les lignes sont contenues dans des balises balises <span style="color:green;">&lt;tr&gt; ... &lt;/tr&gt;</span>
- Les cellules de chaque ligne sont contenues dans des balises balises <span style="color:green;">&lt;td&gt; ... &lt;/td&gt;</span>

En outre, les cellules peuvent être fusionnées verticalement ou horizontalement à l'aide des attributs respectifs <span style="color:#1969a2;">rowspan</span> ou <span style="color:#1969a2;">colspan</span>. 

On notera que les bordures sont configurées à l'aide du `CSS`. En effet, par défaut dans les navigateurs, les tableaux sont affichés sans bordure. Dans l'exemple ci-dessous, elles ont été ajoutées dans un souci de lisibilité.

> ↪ ***Code HTML*** :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Tableau</title>
>         <!-- Bordure des cellules en CSS -->
>         <style>
>             td {border:1px solid lightgray;}
>         </style>
>     </head>
>     <body>
>         <table>
>             <tr>
>                 <td>Row1-Cell1</td>
>                 <td>Row1-Cell2</td>
>                 <td>Row1-Cell3</td>
>                 <td>Row1-Cell4</td>
>             </tr>
>             <tr>
>                 <td rowspan="2">Row2&3-Cell1</td>
>                 <td>Row2-Cell2</td>
>                 <td>Row2-Cell3</td>
>                 <td>Row2-Cell4</td>
>             </tr>
>             <tr>
>                 <td colspan="3">Row3-Cell2&3&4</td>
>             </tr>
>         </table>
>     </body>
> </html>
> ```
> 
> <br >↪ ***Rendu correspondant*** :

<figure>
	<center>
		
		<figcaption></figcaption>
	</center>
</figure>

#### Quelques balises de texte
---

Les balises de texte sont des balises de type `inline` : elles n'interrompent pas le "flot" du texte contrairement aux balises de type `block` qui introduisent un nouveau bloc. Elles permettent principalement d'indiquer la nature du texte.

| Balises | rôle |
| :- | :- |
| <span style="color:green;">&lt;strong&gt; ... &lt;/strong&gt;</span> | Contenu très important |
| <span style="color:green;">&lt;em&gt; ... &lt;/em&gt;</span> | Contenu mis en emphase |
| <span style="color:green;">&lt;mark&gt; ... &lt;/mark&gt;</span> | Texte mis en relief dans un contexte particulier (recherche) |
| <span style="color:green;">&lt;sup&gt; ... &lt;/sup&gt;</span> | Texte en exposant |
| <span style="color:green;">&lt;sub&gt; ... &lt;/sub&gt;</span> | Texte en indice |
| <span style="color:green;">&lt;code&gt; ... &lt;/code&gt;</span> | Texte en police fixe pour écrire du code informatique |

> ↪ ***Code HTML*** :
> ```html
> <!DOCTYPE html>
> <html lang="fr">
>     <head>
>         <meta charset="UTF-8">
>         <title>Balises de texte</title>
>         <!-- Bordure des cellules en CSS -->
>         <style>
>             td {border:1px solid lightgray;}
>         </style>
>     </head>
>     <body>
>         <p>
>             La représentation d'un nombre entier en <strong>binaire</strong> 
>             (ou <strong>base 2</strong>) est une <em>séquence de chiffres compris
>             entre 0 et 1</em>.
>             <br>Exemple : (101)<sub>2</sub> = 
>             <mark>1</mark>* 2<sup>2</sup> + <mark>0</mark>* 2<sup>1</sup> + <mark>1</mark>* 2<sup>0</sup> = 
>             (5)<sub>10</sub>
>         </p>
>         <p>
>             Pour afficher la valeur de (101)<sub>2</sub> en base 10, il suffit d'utiliser la 
>             <strong>notation 0b</strong> comme suit : <code>>>> 0b101</code> 
>         </p>
>     </body>
> </html>
> ```
> <br >↪ ***Rendu correspondant*** :

<figure>
	<center>
		
		<figcaption></figcaption>
	</center>
</figure>

*Remarque* :
> Les balises <span style="color:green;">&lt;strong&gt; ... &lt;/strong&gt;</span>, sont souvent mises en gras par les navigateurs, les balises <span style="color:green;">&lt;em&gt; ... &lt;/em&gt;</span> en italique et les balises <span style="color:green;">&lt;mark&gt; ... &lt;/mark&gt;</span> surlignées, et ce, dans un soucis de lisibilité. Mais la mise en gras, en italique ou le surlignement doivent ***uniquement se faire avec les propriétés `CSS` ad hoc***.
